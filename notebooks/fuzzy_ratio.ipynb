```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
crossref = pd.read_json("../output/crossref_best.json")
crossref.head()
```

```python
s2 = pd.read_json("../output/semantic_scholar_quality.json")
s2.head()
```

```python
def plot_fuzz(
    df: pd.DataFrame, desc: str, col: str = "fuzz_ratio", bins: int = 20
) -> None:
    plt.figure(figsize=(10, 6))
    df[col].hist(bins=bins, edgecolor="black")

    # Add labels and title
    plt.xlabel("Fuzz Ratio")
    plt.ylabel("Frequency")
    plt.title(f"Frequency Distribution of Fuzz Ratio - {desc}")

    # Display the plot
    plt.show()
```

```python
plot_fuzz(crossref, "Crossref")
```

```python
plot_fuzz(s2, "S2")
```

```python
print((s2["fuzz_ratio"] >= 80).value_counts())
print((s2["fuzz_ratio"] >= 80).value_counts(normalize=True))
```
